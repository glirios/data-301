## Flight delays and distance

Create a visualization that explores the following question:

> Are longer/shorter flights more or less on time?

Keep in mind that there are 5000 observations here. To answer this question will likely require some sort of aggregation so draw out the correlations.

```python
# YOUR CODE HERE
Chart(flights).mark_bar().encode(
    X('average(distance)'),
    Y('delay')
)
```

In the following markdown cell discuss conclusions from this visualization in complete sentences:

On average, longer flights tend to show up on time to their destination. The reason for this may be that flights that are longer are given more room in case they hit turbulence and have to find another alternative route that will allow the fight to get to the airport on time. 

## Delays over time

Create a visualization that explores the following question:

> How do average flight delays vary with the day of the week in this dataset?

```python
# YOUR CODE HERE
Chart(flights).mark_point().encode(
    X('date:T', timeUnit='day', title='Day of the Week'),
    Y('average(delay)', title='Average Delay')
)
```

In the following markdown cell discuss conclusions from this visualization in complete sentences:

Based on the graph, the best days to fly are Saturday and Monday because they have the least amount of delays in the week. The worst day to fly out is Friday with an average of about 14 and a half hours of delays.

## Delays by origin and destination

Create a visualization that allows explores the following question:

> Which airports (both origin and detination) have the best and worst average delays?

In creating this visualiztion, there are three variables you need to look at. This means you will need to use at least three encoding channels. Finding the right encoding channels and marks is the challenge here!

```python
# YOUR CODE HERE

Chart(flights, height = 800, width = 800).mark_bar().encode(
    X('origin', sort=SortField(field = 'delay', op = 'average'), title='Destination'),
    Y('destination', sort=SortField(field = 'delay', op = 'average', order = 'descending'), title='Origin'),
    Color('average(delay)', scale= Scale(range=['yellow', 'red']), title='Scale')
)
```

From the bar chart above, TPA to STL and MDW to FLL have averages of delays that seem to be the worst overall when compared to other flights. Additionally, BUF to PHX and LAS, ALB to LAS, and SLC to HOU have the best delay averages. Flights from LAX to other Also, flights from LAS to other airports seems to have higher average delays.
